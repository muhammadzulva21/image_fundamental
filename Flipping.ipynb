```python
import cv2
from google.colab.patches import cv2_imshow
#Membaca Img
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/051.jpg', cv2.IMREAD_UNCHANGED)
```

```python
print("Original Image")
print(cv2_imshow(img))

# flip the image horizontally
print("[INFO] flipping image horizontally...")
flipped = cv2.flip(img, 1)
cv2_imshow(flipped)

# flip the image vertically
flipped = cv2.flip(img, 0)
print("[INFO] flipping image vertically...")
cv2_imshow(flipped)

# flip the image along both axes
flipped = cv2.flip(img, -1)
print("[INFO] flipping image horizontally and vertically...")
cv2_imshow(flipped)
```
