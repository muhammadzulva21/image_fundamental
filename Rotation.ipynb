```python
import cv2
from google.colab.patches import cv2_imshow
#Membaca Img
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/051.jpg', cv2.IMREAD_UNCHANGED)
```

```python
# Rotation img
(h, w) = img.shape[:2]
center = (w/2, h/2)

M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(img, M, (w,h))
cv2_imshow(rotated)
```

```python
M = cv2.getRotationMatrix2D((center), 45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))
cv2_imshow(rotated)
```

```python
M = cv2.getRotationMatrix2D((center), -90, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))
cv2_imshow(rotated)
```
