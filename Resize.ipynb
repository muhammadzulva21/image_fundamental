```python
import cv2
from google.colab.patches import cv2_imshow
#Membaca Img
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/051.jpg', cv2.IMREAD_UNCHANGED)
```

```python
# resize img (abaikan aspek rasio)
im_resized = cv2.resize(img, (400,400))
cv2_imshow(im_resized)

# resize img (mempertahankan aspek rasio)
r = 400/img.shape[1]
dim = (400,int(img.shape[0]*r))
im_resized = cv2.resize(img, dim)
cv2_imshow(im_resized)
```
